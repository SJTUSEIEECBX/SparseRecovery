# Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
```

```python
result = np.load('results/K500Ka50NBS64SNR20sim300AMP.npz')
Pe_p = result['arr_1']
Pe_h = result['arr_4']
T_set = np.arange(30, 82, 2)
plt.semilogy(T_set, Pe_p, '-o')
plt.semilogy(T_set, Pe_h, '--^')
plt.xlim([30, 55])
plt.grid(True, linestyle='-.')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-.', alpha=0.2)
plt.legend(['$P_e^{param}$', '$P_e^{ch}$'])
plt.xlabel('Time Overhead $G/time\ slot$')
plt.ylabel('Error Probability $P_e$')
plt.savefig(fname='Probability_of_errorKa100.pdf')
plt.show()
```
